```python
import os
import numpy as np
import urllib.request as url
import wikipedia as w
from bs4 import BeautifulSoup as bs
```

# Generating the dataset for the AWS Deep Learning hackathon
First, get the list of paintings

```python
folders = os.listdir()
folders = [folder for folder in folders if 'ipynb' not in folder and 'zip' not in folder]
images = [os.listdir(folder) for folder in folders]
images = [im for image in images for im in image]
```

```python
names = [im.split('_')[0] for im in images]
names = np.unique(names)
```

Now get the list of female artists

```python
d = {}
w.set_lang('en')
s = w.page(w.search('Lists of women artists')[0])
html = url.urlopen(s.url).read()
soup = bs(html, 'html.parser')
```

```python
for row in soup.findAll('tr'):
    for colnum, col2 in enumerate(row.find_all('td')):
        for col in col2.find_all('li'):
            a = col.find('a')
            link = 'https://en.wikipedia.org' + a.get('href')
            d[a.get('title')] = {}
            d[a.get('title')]['link'] = link
```

```python
artists = list()
for country, link in d.items():
    page = bs(url.urlopen(link['link']).read(), 'html.parser')
    for ul in page.find_all('ul', {'class':"vector-menu-content-list"}):
        ul.decompose()
    for ul in page.find_all('ul', {'class':"noprint"}):
        ul.decompose()    
    for ul in page.find_all('ul', {'id':"footer-info"}):
        ul.decompose()    
    for ul in page.find_all('ul', {'id':"footer-places"}):
        ul.decompose()    
    for ul in page.find_all('ul'):
        for li in ul.find_all('li'):
            aa = li.find_all('a')        
            for a in aa:            
                tmp = a.get('title')
                if not tmp is None:
                    if 'Category' not in tmp:
                        artists.append(tmp)
```

```python
len(artists)
```

```python
artists = [x for x in artists if not "List of" in x]
artists = [x for x in artists if not "Template" in x]
artists = [artist.replace('(artist)', '').strip() for artist in artists]
artists = [artist.lower().replace(' ', '-') for artist in artists]
```

```python
len(artists)
```

Now filter the images for female artists

```python
female_artists = np.intersect1d(names, artists)
```

```python
female_artists_images = [image for image in images if image.split('_')[0] in female_artists]
female_artists_images
```
